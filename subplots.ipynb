#Sub Plots


```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,4,9,16,25,36,49,64,81]

fig=plt.figure(figsize=(18,5))

first_plot=fig.add_subplot(1,3,1)
first_plot.plot(x,y,color='#ff0080')
first_plot.set_title("Line Plot")
first_plot.grid()

second_plot=fig.add_subplot(1,3,2)
second_plot.scatter(x,y,color='#00ff00')
second_plot.set_title("Scaetter Plot")

third_plot=fig.add_subplot(1,3,3)
third_plot.bar(x,y,color='#00ffff')
third_plot.set_title("Bar Graph")

plt.show()
```
